* **Text Hierarchy in NLP**  
***Paragraph → Sentences → Words → Letters***  

* **Basic Meaning**  
***Words carry the basic meaning of text***  

* **Step 1: Sentence Tokenization**  
***Converting a paragraph into a list of sentences is called sentence tokenization***  

* **How do we detect sentences?**  
***Using punctuation marks***  

* **Punctuation used:**  
***. (full stop), ! (exclamation), ? (question mark)***  

* **Example:**  
***I'm Rixz. Also Known As Batman.***  

* **Step 2: Word Tokenization**  
***After getting sentences, we break each sentence into words → this is called word tokenization***  

* **Do all words carry the same meaning?**  
***No, not all words in a paragraph or sentence carry the same meaning***  

* **Stop Words**  
***Words that do not carry much meaning are called stop words***  

* **Examples:**  
***The, are, a, is, etc.***  

* **Why remove stop words?**  
***To focus on meaningful words and improve model performance***  

* **Who decides stop words?**  
***Defined using a corpus → a rulebook/list containing stop words***  

* **Step 3: Removing Punctuation**  
***After tokenization, we remove punctuation marks from text***  

* **Why remove punctuation?**  
***Punctuation does not add significant meaning for most NLP tasks***  

* **Punctuation removed:**  
***. , ! ? : ; ' " ( ) etc.***  

* **Example:**  
***"Hello, world!" → "Hello world"***  

* **Step 4: Getting Root Words (Stemming / Lemmatization)**  
***Converting words into their root/base form is called stemming or lemmatization***  

* **Why do this?**  
***To reduce different forms of a word into a single base form and improve consistency***  

* **Stemming**  
***Cuts words to root form (may not be a real word)***  
***Example: playing → play, studies → studi***  

* **Lemmatization**  
***Converts words into meaningful base form using vocabulary***  
***Example: playing → play, better → good***  

* **Example where Stemming fails**  
***He ran → run (Stemming fails because it may not correctly convert "ran" to "run")***  

* **Lemmatization**  
***It is a rule-based approach where a vocabulary (corpus) is used to map words to their root forms***  

* **Corpus**  
***A rulebook or dictionary containing base/root words***  

* **Examples:**  
***saw (verb) → see***  
***saw (noun) → a cutting tool***
