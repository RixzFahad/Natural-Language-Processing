***Remove The StopWords***

*   Stemming :- Crude Approch
*   Lemmatization :- Rule Based Approach



```python
from nltk.corpus import stopwords
import string
```

```python
#Stemming Is A Crude Approach Where It Not Works Always Becasue It Removes Last Letter Like Ing, Ed, Er Etc
#Mostly Stemming Won't Works With Simple Past Text Data.
#To Avoid This Problem We Use The Lemmatization.
#Lemmatization Is A Rule Based Approach To Get The Root Words.
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
stop_words = set(stopwords.words('english'))
print(stop_words)
#Initializing The Stemming And Lemmatizer
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

for w in sentence:
    words = nltk.word_tokenize(w)

    # remove punctuations
    words = [word for word in words if word not in string.punctuation]

    # remove stop words
    words = [word for word in words if word.lower() not in stop_words]

    # get root words
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    stemming_words = [stemmer.stem(word) for word in words]

    print("Original word:", w)
    print("Lemmatized words:", lemmatized_words)
    print("Stemming words:", stemming_words)
    print()
```

```python
document = ['Good Movie good Screenplay', 'good movie','bad movie sleep']
# import libraries To Convert text data into numeric forms
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
```

***Step - One*** **Design The Vocabulary**

```python
count_vectorizer = CountVectorizer(binary=False)
```

***Step - Two*** **Create A bag Of Words Model**

```python
bag_of_words = count_vectorizer.fit_transform(document)
```

***Step - Three*** **Show The Bag Of Words As A Pandas DF**

```python
feature_name = count_vectorizer.get_feature_names_out()
pd.DataFrame(bag_of_words.toarray(), columns=feature_name)
```

***N-Gram Approach*** - **Bi-Gram Approach**

```python
document  = ['good movie good screenplay','good movie','bad movie sleep']
# import library to covert text dat into numerical forms
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# step 1:- Design the vocabulary
#ngram_range = (2,2) means only bigram
count_vectorizer = CountVectorizer(ngram_range=(2,2))
# step 2:- create a bag of wrods model
bag_of_words = count_vectorizer.fit_transform(document)
# step 3 :- show the bag of words as a pandas dataframe
feature_name = count_vectorizer.get_feature_names_out()
pd.DataFrame(bag_of_words.toarray(),columns=feature_name)
```

***TF-IDF***

```python
document  = ['good movie good screenplay','good movie','bad movie sleep']
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
tfidf_vectorizer = TfidfVectorizer()
values = tfidf_vectorizer.fit_transform(document)
feature_names = tfidf_vectorizer.get_feature_names_out()
pd.DataFrame(values.toarray(),columns=feature_names)


#Inverse Document Frequency- Term Frequency
```
