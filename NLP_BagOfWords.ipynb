# 📘 Bag of Words (BoW)

**Definition:**  
Bag of Words is a technique used to convert text data into numerical form so that it can be used in machine learning models.

**Key Idea:**  
- It counts how many times each word appears in a sentence/document.  
- It ignores grammar and word order — only focuses on frequency.

---

## 🔹 Why is it used?
- To represent text data as numbers  
- Helps machine learning models understand text  
- Simple and effective baseline for NLP tasks

---

## 🔹 Approaches to Convert Text into Numerical Form

### 1. Setting up the Vocabulary
- Create a list of all unique words in the dataset  
- Each word becomes a feature (column)  
- Count occurrences of each word in a sentence  

**Example:**  
Sentence: *"I love AI"*  
Vocabulary: `[I, love, AI]`  
Vector: `[1, 1, 1]`

---

### 2. N-gram Approach
- Instead of single words, consider sequences of words  

**Types:**
- Unigram → single words → *"I", "love", "AI"*  
- Bigram → pairs → *"I love", "love AI"*  
- Trigram → triplets → *"I love AI"*

**Purpose:**  
- Captures context better than single words  

---

## ⚡ Summary
- BoW = Text → Numbers (by counting words)  
- Two main approaches:
  1. Vocabulary-based (word count)
  2. N-grams (word sequences)
